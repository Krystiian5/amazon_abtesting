### Limpieza Optimizada
Aqui vamos a ir optimizando las funciones (simplificando y resumiendo)

```python
import pandas as pd

#Exploracion de archivos
import sys
sys.path.append('../')  # asegúrate de apuntar al root del proyecto
#Funciones personalizadas
import src.sp_limpieza as sl
```

```python
df_raw = pd.read_csv('../data/data_raw.csv')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = df_raw.copy()
```

### Cambiar nombre de las columnas

```python
df.columns = df.columns.str.replace(' ', '_').str.lower()
```

### Cambiar visit_date a fecha

```python
df['visit_date'] = pd.to_datetime(df['visit_date'], format = '%Y-%m-%d')
```

### Conversion
Vamos a crear una nueva columna, conversion str

```python
df['conversion_str'] = df['conversion'].apply(lambda x: 'yes' if x == 1 else 'no')
```

### Cambiar Product Category 'Fashhion' a 'fashion'

```python
df['product_category'] = df['product_category'].str.replace('fashhion', 'fashion')
```

### Product Id Cambiar a numérica

```python
#Quitamos la P y luego modificamos el tipo de columna

df['product_id'] = df['product_id'].str.replace('P','')
df['product_id'] = df['product_id'].astype('int64')
```

### Browser cambiar las categorias que estan mal

```python
dicc_map = {'safaaaaari' : 'safari',
            'ottttther' : 'other',
            'firefoxxxxxx' : 'firefox',
            'eddddge' : 'edge',
            'chrooome' : 'chrome'}
```
