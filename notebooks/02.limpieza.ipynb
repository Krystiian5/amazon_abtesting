```python
import pandas as pd
```

```python
df_raw = pd.read_csv('../data/data_raw.csv')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = df_raw.copy()
```

### Cambiar nombre de las columnas

```python
df.columns = df.columns.str.replace(' ', '_').str.lower()
```

```python
df.columns
```

### Cambiar visit_date a fecha

```python
df.sample(5)
```

```python
df['visit_date'] = pd.to_datetime(df['visit_date'], format = '%Y-%m-%d')
```

### Conversion
Vamos a crear una nueva columna, conversion str

```python
df['conversion_str'] = df['conversion'].apply(lambda x: 'yes' if x == 1 else 'no')
```

```python
df.loc[:,['conversion','conversion_str']].sample(5)
```

### Cambiar Prodduct Category a minúsculas. Cambiar 'Fashhion' a 'fashion'

```python
df['product_category'] = df['product_category'].str.lower()
```

```python
df['product_category'] = df['product_category'].str.replace('fashhion', 'fashion')
```

```python
df['product_category'].unique()
```

### Product Id Cambiar a numérica

```python
#Quitamos la P y luego modificamos el tipo de columna

df['product_id'] = df['product_id'].str.replace('P','')
df['product_id'] = df['product_id'].astype('int64')
```

```python
df.sample()
```

```python
df.info()
```

### Product name. Cambiar ' ' por '_' y a minusculas

```python
df['product_name'] = df['product_name'].str.replace(' ','_')
```

Aqui vemos que tenemos que volver a utilizar el codigo para hacerlas minusculas. Vamos a crear una funcion para no tener que repetir el codigo

```python
def minus(df, col):
    df[col] = df[col].str.lower()
```

```python
minus (df,'product_name')
```

```python
df.sample(5)
```

### Price, discount y total value Cambiar ',' por '.' y cambiar a tipo float.
Igual vamos a crear una funcion porque la vamos a utilizar mas veces

```python
def comas (df,col):
    df[col] = df[col].str.replace(',','.')
    df[col] = df[col].astype('float64')
```

```python
comas (df, 'price')
```

```python
comas (df, 'discount')
comas (df, 'total_value')
```

```python
df.sample(5)
```

```python
df.info()
```

### Payment Method y shipping method Cambiar ' ' por '_'
Creamos otra funcion

```python
def espacios(df, col):
    df[col] = df[col].str.replace(' ', '_')
```

```python
espacios (df, 'payment_method')
espacios (df, 'shipping_method')
minus (df, 'payment_method')
minus (df, 'shipping_method')
```

### Cambiar Device, browser y Referral a minus

```python
minus (df, 'device')
minus (df, 'referral_source')
minus (df, 'browser')
```
